```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("IRIS.csv")
print(df.head())
```

```python
# plotting strip plot with seaborn
ax=sns.stripplot(data=df, x="species", y="sepal_width")

#giving labels to x-axis nand y axis
ax.set(xlabel= "Iris Species", ylabel="sepal Width")
#use to set style of background of plot
sns.set(style="whitegrid")

#givingtitle to the plot
plt.title("Strip Plot in Python")

#function to show plot
plt.show()
```

```python
#plotting swarm plot with seaborn
ax=sns.stripplot(data=df, x="species", y="sepal_width")

#giving labels to x-axis nand y axis
ax.set(xlabel= "Iris Species", ylabel="sepal Width")

#use to set stule of background of plot
sns.set(style="whitegrid")

#giving title to the plot
plt.title("Swarm Plot in python")

#function to show plot
plt.show()
```
